```python
import pandas as pd
```

```python
df = pd.read_csv('healthcare_sgx.csv')
```

```python
df.set_index(['company','year'],inplace=True)
```

```python
# to see what is all object types for only this line
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  
    print(df.dtypes[df.dtypes=='object'])
```

```python
# drop useless row
df.drop([i for i in df.columns if 'Ended' in i]+[i for i in df.columns if 'Source' in i],axis=1,inplace=True)
```

```python
# change period length
df['Period Length'] = [i.split(' ')[0] for i in df['Period Length']]
```

```python
df.replace(',','', regex=True, inplace=True)
df = df.apply(pd.to_numeric, errors='coerce')
```

```python
df.dtypes[df.dtypes=='object']
```

```python
# keeping only ratios
ratios = ['Return on Assets (ROA)','Return on Equity (ROE)','Operating Margin',
                  'Operating Margin','Net Profit Margin','Asset Turnover',
                  'Current Ratio','Quick Ratio','Debt / Equity']
```

```python
health_ratios = df[ratios]
```
